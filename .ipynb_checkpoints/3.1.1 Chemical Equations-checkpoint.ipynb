## `mhchem`

Typesetting of chemical equations in a markdown cell is possible using the `mhchem` *MathJax* plugin ([`mhchem/MathJax-mhchem](https://github.com/mhchem/MathJax-mhchem)).

To enable the plugin, `$$\require{mhchem}$$` needs adding to - and running? - in a markdown cell' `mhchem` style commands should then be available in other markdown cells.

## `chemfig`

The [`chemfig` *TeX* package](http://tug.ctan.org/macros/generic/chemfig/chemfig-en.pdf) provides support for drawing a wide range of molecules as 2D images that are appropriate for print as well as online display.

See [wikibooks/LaTeX/Chemical_Graphics](https://en.wikibooks.org/wiki/LaTeX/Chemical_Graphics) for plenty of hints about using `chemfig`.

```python
%%tikz  -p chemfig --no-wrap -s 0.4

\chemfig{C(-[:0]H)(-[:90]H)(-[:180]H)(-[:270]H)}
```
