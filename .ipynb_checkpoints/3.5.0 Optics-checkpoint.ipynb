# Optics

Resources that are useful for creating eucational materials related to the study of optics.

At first glance, these all look a bit complicated, but there may be some easy ways in to simple mirror and lens diagrams with the `rayopt` package, which can do ray trace diagrams of simple setups, for example. 

## `LightPipes`

[`LightPipes`](https://github.com/opticspy/lightpipes) models coherent optical devices when the diffraction is essential.

```python
%%capture
!pip install LightPipes
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

import LightPipes as lp
from LightPipes import um, mm, cm

wavelength=5*um
size=20.0*mm
N=500
z=50*cm
R=0.3*mm
d=1.2*mm

F=lp.Begin(size,wavelength,N)
F1=lp.CircAperture(R/2.0,-d/2.0, 0, F)
F2=lp.CircAperture(R/2.0, d/2.0, 0, F)    
F=lp.BeamMix(F1,F2)
F=lp.Fresnel(z,F)
I=lp.Intensity(2,F)
plt.imshow(I, cmap='rainbow'); plt.axis('off');plt.title('intensity pattern')
plt.show()
```

## `rayopt`

[`rayopt`](https://github.com/quartiq/rayopt) is a Python package for optics design (lenses, cavities, gaussian optics, lasers) capable of generating associated ray tracing diagrams.

A wide range of example notebooks can be found at [quartiq/rayopt-notebooks](https://github.com/quartiq/rayopt-notebooks), although they require data from third party sources to be loaded in to a materials database (and it's not obvious how to get the original datafiles or load them into the local db).

```python
%%capture
!pip3.6 install rayopt
```

```python
import numpy as np
import matplotlib.pyplot as plt

import rayopt as ro

# ignore matplotlib and numpy warning each other
import warnings
warnings.simplefilter("ignore", FutureWarning)
# ignore floating point exceptions
np.seterr(divide="ignore", invalid="ignore")
# by default only print 4 significant digits
np.set_printoptions(precision=4)
```

```python
s = ro.system_from_yaml("""
object:
  pupil:
    radius: 1
elements:
- {}
- {distance: 1, material: 1.5, roc: 5, radius: 1}
- {distance: .2, material: 1.0}
- {}
""")
s.update()
print(s)
```

```python
ro.Analysis(s);
```

```python
#This requires the analysis to have been run previously - presumably it updates the s object
s.object.pupil.radius = .1
q = ro.GaussianTrace(s)
fig, ax = plt.subplots()
s.plot(ax)
q.plot(ax, color="red", scale=10)
s.object.pupil.radius = 1.
```

### Hubble Telescope

```python
#https://github.com/quartiq/rayopt-notebooks/blob/8600ac95abe8e932fc5bdded171c35046af1d2db/hubble_space_telescope.ipynb
from rayopt import System, Spheroid, InfiniteConjugate, Analysis, vacuum, mirror

s = System(description="Hubble Space Telescope 2.4m f/24",
          wavelengths=[632.8e-9])
s.object = InfiniteConjugate(angle_deg=154/60**2/2**.5)
s.fields = 0, .7, 1.
start = Spheroid(material=vacuum)
m1 = Spheroid(distance=5e3, material=mirror)
m2 = Spheroid(radius=150., material=mirror)
end = Spheroid()
s.extend([start, m1, m2, end])

# Ritchey-Chrétien telescope
f = 57599.
d = 4906.
b = d + 1500
m1.radius = f/24/2
s.object.pupil.radius = m1.radius
m1.curvature = -(f - b)/(2*d*f)
m2.curvature = (f - d - b)/(2*d*b)
m = (f - b)/d
m1.conic = -1 - 2/m**3*b/d
m2.conic = -1 - 2/(m - 1)**3*(m*(2*m - 1) + b/d)
m2.offset = 0, 0, -d
end.distance = b
end.curvature = 2*(m1.curvature - m2.curvature)  # field curvature
end.curvature *= .96
s.update()
```

```python
a = Analysis(s)
```

```python
columns = "type roc distance radius material"
text = """
    O       0      0 .05  AIR
    S  .55622     .1 .033 glass/SCHOTT-SK|N-SK4
    S -.15868 .00632 .033 AIR
    S  .03982 .00061 .028 glass/SCHOTT-SK|N-SK4
    S .089438 .01464 .023 AIR
    S -.07858 .01008 .018 glass/SCHOTT-SF|N-SF5
    S  .03237 .00254 .016 AIR
    A       0 .01632 .013 AIR
    S   .1191 .00056 .015 glass/SCHOTT-SK|N-SK4
    S -.04214 .00442 .015 AIR
    I       0  .0735 .05  AIR
    """

s = ro.system_from_text(text, columns.split(), scale=1.,
    description="four element double gauss, intermediate optical design")
s.rescale()
s.object.angle = np.deg2rad(15)
s.object.pupil.radius = 100/2.8/2
s.fields = 0, .7, 1.
s.update()
```

```python
a = ro.Analysis(s)
```

## `POPPY`

The [`POPPY` (Physical Optics Propagation in PYthon)](http://poppy-optics.readthedocs.io/en/stable/index.html) package  simulates physical optical propagation including diffraction.

```python
%%capture
!pip3.6 install poppy 
```

```python
import poppy

oversample=2
pixelscale = 0.010  #arcsec/pixel
wavelength = 4.6e-6

osys = poppy.OpticalSystem("test", oversample=oversample)
osys.add_pupil(poppy.CircularAperture(radius=6.5/2))
osys.add_image()
osys.add_image(poppy.BandLimitedCoron(kind='circular',  sigma=5.0))
osys.add_pupil()
osys.add_pupil(poppy.CircularAperture(radius=6.5/2))
osys.add_detector(pixelscale=pixelscale, fov_arcsec=3.0)

osys.source_offset_theta = 45.
osys.source_offset_r =  0.1  # arcsec
psf = osys.calc_psf(wavelength=wavelength, display_intermediates=True)
```

## `prysm`

[`prysm`](https://github.com/brandondube/prysm) is a general optical modeling and analysis library.

<blockquote> The term pupil in prysm refers to the pupil of an optical system, not e.g. the pupil of the human eye. Poppy and prysm are somewhat sister libraries, the core of prysm is numerical optical propagation, which is all Poppy does. prysm does many other things, while poppy offers much more flexible propagation (prysm only does fresnel transforms, I think poppy also does angular spectrum and all of the others). The purpose of poppy is similar to that of proper (another similar lib) -- to do multi-plane diffraction simulations.

Prysm is a more general optical modeling and analysis library.

Via: https://github.com/brandondube/prysm/issues/1#issuecomment-409756462</blockquote>

```python
%%capture
#Also needs py 3.6 kernel - support for f-strings required
!pip3.6 install prysm
```

```python
import numpy as np
from prysm import Pupil

# create a pupil object
# 5mm pupil with light of wavelength 1 micron and 512x512 samples
p = Pupil(epd=5, wavelength=1, samples=512) 
p.plot2d();
```

```python
# add a decaying sine wave in x and sine wave in y of phase error.
x, y = np.linspace(-1,1, p.samples), np.linspace(-1,1,p.samples)
xx, yy = np.meshgrid(x,y)
p.phase = np.cos(xx*1.5) * np.sin(yy*1.5)

#The following throws an error - the .clip() method doesn't exist
#p.clip();  # if you modify the phase, you should call .clip to re-enforce the mask
#p._phase_to_wavefunction() suggested as replacement but doesn't have same effect as .clip()?
# pupils can be plotted in 2D
p.plot2d()
```

```python
# simulated interferograms
p.interferogram()

# can specify number of passes and visibility
p.interferogram(passes=4, visibility=0.5);
```
