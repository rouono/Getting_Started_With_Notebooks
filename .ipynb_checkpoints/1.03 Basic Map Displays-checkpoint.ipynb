# Basic Map Displays

A wide variety of packages exist for working with and rendering geospatial data and various map representations as well as embedding interactive web maps in a Jupyter notebook.

This notebook is a working document for exploring various packages that can be used to generate static and interactive maps.

The intention is two-fold, to explore:

- the extent to which notebook compatible packages can be used to support the creation of cartographic resources for use in educational materials (and by extension whether simplifying the production process makes the use of maps more attractive when producing educational resources); and
- the extent to which providing interactive maps, as well as the means of producing maps, allows a learner to engage more deeply with cartographic resources.

(I think this is in part what motivated OpenCreate: making the means of production for creating resources available to students so that they could create their own resources as a way of exploring a topic.)

```python
m = Basemap(width=12000000,height=9000000,projection='lcc',
            resolution='c',lat_1=45.,lat_2=55,lat_0=50,lon_0=-107.)
# draw coastlines.
m.drawcoastlines()
# draw a boundary around the map, fill the background.
# this background will end up being the ocean color, since
# the continents will be drawn on top.
m.drawmapboundary(fill_color='aqua')
# fill continents, set lake color same as ocean color.
m.fillcontinents(color='coral',lake_color='aqua')
plt.show()
```

```python
# setup Lambert Conformal basemap.
# set resolution=None to skip processing of boundary datasets.
m = Basemap(width=12000000,height=9000000,projection='lcc',
            resolution=None,lat_1=45.,lat_2=55,lat_0=50,lon_0=-107.)
m.shadedrelief()
plt.show()
```

### Pulling Map data into Basemap

We can pull map imagery into `basemap` using the [*arcgisimage*](https://basemaptutorial.readthedocs.io/en/latest/backgrounds.html) API.


See also http://qingkaikong.blogspot.com/2016/06/nice-python-basemap-background.html  - eg setting pixel dimansions.

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt


#IW 50.6938 N, 1.3047 W, 50.6938, -1.3047
iwlat_0 = 50.6938
iwlon_0 = -1.3047

iwllcrnrlon=-1.65
iwllcrnrlat=50.5
iwurcrnrlon=-1.
iwurcrnrlat=50.9

epsg= 27700
```

```python
lat_0 = iwlat_0
lon_0 = iwlon_0

llcrnrlon=iwllcrnrlon
llcrnrlat=iwllcrnrlat
urcrnrlon=iwurcrnrlon
urcrnrlat=iwurcrnrlat
```

```python
demomap = Basemap(llcrnrlon=llcrnrlon,llcrnrlat=llcrnrlat,
                  urcrnrlon=urcrnrlon,urcrnrlat=urcrnrlat, epsg=epsg)
#web mercator: epsg:4236, OSGB epsg:27700
#http://server.arcgisonline.com/arcgis/rest/services

demomap.arcgisimage(service='ESRI_Imagery_World_2D', xpixels = 1500, verbose= True)
plt.show()
```

```python
demomap.arcgisimage(service='World_Shaded_Relief', xpixels = 1500, verbose= True)
plt.show()
```

```python
from ipywidgets import interact

#Some of this may not work....
#the list is from http://server.arcgisonline.com/arcgis/rest/services
maptypes=["ESRI_Imagery_World_2D", "ESRI_StreetMap_World_2D", "I3_Imagery_Prime_World",
          "NASA_CloudCover_World", "NatGeo_World_Map", "NGS_Topo_US_2D", "Ocean_Basemap",
          "USA_Topo_Maps","World_Imagery", "World_Physical_Map", "World_Shaded_Relief",
          "World_Street_Map","World_Terrain_Base","World_Topo_Map"]

@interact(maptype=maptypes, title='Demo Map')
def esriMaps(maptype):
    demomap.arcgisimage(service=maptype, xpixels = 1500, verbose= True)
    plt.show();
```

## Create Inset Maps

One the the things we are likely to want to do in many educational materials is to create inset maps.

If we simplify the process of creating such maps within our educational resource authoring environment, are we more likely to make use of them?

For example:

```python
from operator import itemgetter

from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
import numpy as np


def createInsetMap(base=None, inset=None,
                   zoom=None, loc=1, baseresolution=None,
                   insetresolution=None, connector=True):

    ''' Create an inset map for a particular region. '''

    fig = plt.figure()
    ax = fig.add_subplot(111)
    
    #Create a basemap
    
    #Add some default logic to provide some demos
    #If no args, use a World map
    if base is None or base.lower=='world':
        baseargs={'projection':'cyl','lat_0':0, 'lon_0':0}
    #If set UK base, then we can add a more detailed inset...
    elif isinstance(base,str) and base.lower()=='uk':
        baseargs={'llcrnrlon':ukllcrnrlon,'llcrnrlat':ukllcrnrlat,
                  'urcrnrlon':ukurcrnrlon,'urcrnrlat':ukurcrnrlat,
                  'resolution':'l'}
    else:
        #should really check base is a dict
        baseargs=base
    
    if baseresolution is not None:
        baseargs.update({'resolution':baseresolution})
        
    map1=Basemap(**baseargs)

    map1.drawmapboundary(fill_color='#f8f8ff')
    map1.drawcoastlines()

    plt.xticks(visible=False)
    plt.yticks(visible=False)
    
    
    #Now define the inset map

    #This default is to make for some nice default demos
    #With no explicit settings, inset UK on World map
    if base is None and inset is None:
        insetargs={'llcrnrlon':ukllcrnrlon,'llcrnrlat':ukllcrnrlat,
                  'urcrnrlon':ukurcrnrlon,'urcrnrlat':ukurcrnrlat,
                  'resolution':'l'}
        zoom=10
        loc=3
    #If the base is UK, and no inset is selected, demo an IW inset
    elif (isinstance(base,str) and base.lower()=='uk') and inset is None:
        insetargs={'llcrnrlon':iwllcrnrlon,'llcrnrlat':iwllcrnrlat,
                   'urcrnrlon':iwurcrnrlon,'urcrnrlat':iwurcrnrlat,
                   'resolution':'h'}
        zoom=10
        loc=2
    else:
        #Should really check inset is a dict...
        insetargs=inset

    axins = zoomed_inset_axes(ax, zoom, loc=loc)
    
    #The following seem to be set automatically?
    #axins.set_xlim(llcrnrlon, llcrnrlat)
    #axins.set_ylim(llcrnrlat, urcrnrlat)

    if insetresolution is not None:
        insetargs.update({'resolution':insetresolution})

    map2 = Basemap(ax=axins,**insetargs)
    
    #map2.drawmapboundary(fill_color='#7777ff')
    map2.fillcontinents(color='#ddaabb', lake_color='#7777ff', zorder=0)
    map2.drawcoastlines()
    map2.drawcountries()

    #Add connector lines from marked area on large map to zoomed area
    if connector:
        mark_inset(ax, axins, loc1=2, loc2=4, fc="none", ec="0.5")

    plt.show()
```

```python
createInsetMap()
```

```python
createInsetMap(base='uk')
```

```python
#Would be good to have a function that could set rough areas around a particular location
#eg relative lat/lon from centre lat/lon?
iwinset={'llcrnrlon':iwllcrnrlon,'llcrnrlat':iwllcrnrlat,
         'urcrnrlon':iwurcrnrlon, 'urcrnrlat':iwurcrnrlat}

createInsetMap(inset=iwinset,zoom=100, loc=4, insetresolution='i')
```

# `folium` Interactive Web Map Embedding

The `folium` package provides a simple Python package for embedding interactie web maps in a Jupyter notebook using the `leaflet` Javascript package. `folium` can also be used to export a standalone html file containing all that's required to embed the map, using an iframe, in another web page. 

`folium` is not installed in the Azure Notebooks environment by default, so we need to install it ourselves. 

```python
%%capture
!pip install folium
```

One we import the package we can use it to create and embed an interactive map centered on a specified latitue/longitude and displayed at a particular zoom level.

```python
import folium
```

```python
folium.Map(location=[52.0250, -0.7084], zoom_start=16)
```

### Different Map Tiles

Maps can be rendered using a wide variety of free map tiles without any API key requirements, including:

- *OpenStreetMap* (default)
- *Mapbox* (*Bright* and *Control Room*) - limited levels of zoom for free tiles
- *Stamen* (*Terrain*, *Toner* and *Watercolor*)
- *CartoDB* (*positron* and *dark_matter*)

Set the `tiles` attribute when creating a map to use a speicifc (supported) tileset.

```python
folium.Map(location=[52.0250, -0.7084], zoom_start=16, tiles='stamentoner')
```
